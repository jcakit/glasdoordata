```python
df['Company Name'] = df['Company Name'].apply(lambda x: x.split('\n')[0] if '\n' in x else x)
df['Salary Estimate'] = df['Salary Estimate'].str.extract('(\d+K-\$?\d+K)').astype(str)

# Splitting the 'Salary Estimate' into two columns: 'Min Salary' and 'Max Salary'
salary_split = df['Salary Estimate'].str.split('-', n=1, expand=True)
df['Min Salary (K)'] = salary_split[0].str.extract('(\d+K)').replace('K','',regex=True).astype(float)
df['Max Salary (K)'] = salary_split[1].str.extract('(\d+K)').replace('K','',regex=True).astype(float)
# Dropping the original 'Salary Estimate' as it's redundant now
df.drop('Salary Estimate', axis=1, inplace=True)

df.head()
```

```python
# Converting 'Founded' to a string since it's a year and we don't perform numerical operations on it
df['Founded'] = df['Founded'].astype(str)
df.dtypes
```

```python
# Function to extract state from the location column assuming the format is "City, State"
def extract_state(location):
    # Split the location by comma and take the second part as state if it exists
    parts = location.split(',')
    if len(parts) > 1 and len(parts[1].strip()) == 2:  # Check if it looks like a state abbreviation
        return parts[1].strip()
    return None  # Return None if no state abbreviation is found

# Apply the function to the location column to create a new state column
df['State'] = df['Location'].apply(extract_state)

# Check the first few entries of the new state column
df[['Location', 'State']].head()
```

```python
# Define keywords that might indicate remote work
remote_keywords = ['remote', 'work from home', 'work at home', 'virtual', 'telecommute', 'distributed team']

# Function to determine if job description mentions remote work
def indicates_remote_work(description):
    description = description.lower()  # Convert to lowercase for matching
    for keyword in remote_keywords:
        if keyword in description:
            return 1  # Indicate remote work is offered
    return 0  # Indicate remote work is not mentioned

# Apply the function to create a new binary column indicating remote work
df['Remote'] = df['Job Description'].apply(indicates_remote_work)

# Check how many job descriptions mention remote work
remote_work_count = df['Remote'].sum()
remote_work_count, df[['Job Description', 'Remote']].head()
```

```python
# Convert the binary 'remote' feature to 'Yes'/'No'
df['Remote'] = df['Remote'].map({1: 'Yes', 0: 'No'})

# Check the updated 'remote' column
df[['Job Description', 'Remote']].head()
```

```python
# Define seniority level keywords
seniority_keywords = {
    'Senior': ['senior', 'sr', 'lead', 'principal', 'manager', 'director', 'head'],
    'Junior': ['junior', 'jr', 'associate', 'assistant', 'intern']
}

# Function to categorize job posting by seniority level based on job title
def categorize_seniority(job_title):
    job_title_lower = job_title.lower()
    for seniority, keywords in seniority_keywords.items():
        if any(keyword in job_title_lower for keyword in keywords):
            return seniority
    return 'Mid'  # Default category

# Apply the function to create a new 'seniority_level' feature
df['Seniority Level'] = df['Job Title'].apply(categorize_seniority)

# Check the distribution of the new 'seniority_level' feature
seniority_level_distribution = df['Seniority Level'].value_counts()
seniority_level_distribution, df[['Job Title', 'Seniority Level']].head()
```
